```python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create an interactive line plot with a slider for amplitude
amplitude_slider = widgets.FloatSlider(value=1.0, min=0.1, max=2.0, step=0.1, description='Amplitude:')
output_plot = widgets.Output()

def update_plot(change):
    with output_plot:
        plt.figure()
        plt.plot(x, change.new * y)
        plt.xlabel('X-axis')
        plt.ylabel('Y-axis')
        plt.title('Interactive Matplotlib Plot with Amplitude Slider')
        plt.grid(True)
        plt.show()

amplitude_slider.observe(update_plot, 'value')
display(amplitude_slider, output_plot)
```

```python
import ipyvolume as ipv
import numpy as np

# Generate 3D data
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create an interactive 3D surface plot using ipyvolume
fig = ipv.figure()
ipv.plot_surface(X, Y, Z, color="orange")
ipv.xlabel("X-axis")
ipv.ylabel("Y-axis")
ipv.zlabel("Z-axis")
ipv.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as widgets
from IPython.display import display

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create an interactive line plot with widgets for amplitude and frequency
amplitude_slider = widgets.FloatSlider(value=1.0, min=0.1, max=2.0, step=0.1, description='Amplitude:')
frequency_slider = widgets.FloatSlider(value=1.0, min=0.1, max=5.0, step=0.1, description='Frequency:')
output_plot = widgets.Output()

def update_plot(change):
    with output_plot:
        plt.figure()
        plt.plot(x, change.new[0] * np.sin(change.new[1] * x))
        plt.xlabel('X-axis')
        plt.ylabel('Y-axis')
        plt.title('Interactive Matplotlib Plot with Widgets')
        plt.grid(True)
        plt.show()

widgets.interactive(update_plot, amplitude=amplitude_slider, frequency=frequency_slider)
display(output_plot)
```
